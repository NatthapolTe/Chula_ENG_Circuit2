```python
import matplotlib.pyplot as plt 
%matplotlib inline
import pandas as pd 
import numpy as np 

filename = 'acq0001.csv'
df = pd.read_csv('acq0001.csv', skiprows=19)
t = df['Time (s)'] - df['Time (s)'][0]
Vs = df['Channel 1 (V)']
Vc = df['Channel 2 (V)']
    
plt.plot(t, Vs, label = "V_s")
plt.plot(t, Vc, label = "V_c")
plt.plot(t, Vs-Vc, label = "V_R")
plt.legend()
plt.xlabel('t (s)') 
plt.ylabel('V (V)')
plt.xlim([0, 0.05]) #เปลี่ยนเองตามใจ
plt.grid()
plt.show()
```
